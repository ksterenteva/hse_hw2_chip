```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_path = "annotated_peaks.txt"
columns = ["PeakID", "Chr", "Start", "End", "Strand", "Annotation", "Detailed_Annotation", "Distance_to_TSS"]
df = pd.read_csv(file_path, sep="\t", comment="#", header=None, names=columns)

# Подсчет количества пиков в разных аннотациях
annotation_counts = df["Annotation"].value_counts()

# Визуализация распределения
plt.figure(figsize=(12, 6))
sns.barplot(x=annotation_counts.index, y=annotation_counts.values)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Геномная аннотация")
plt.ylabel("Количество пиков")
plt.title("Распределение воспроизводимых пиков по геному")
plt.show()
```
